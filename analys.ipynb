```python
import pandas as pd 
from pubmed_lookup import PubMedLookup
from pubmed_lookup import Publication
```

```python
codelund=pd.read_csv('./output/codesharing_lund.csv')
restlund=pd.read_csv('./output/resttransparency_lund.csv')

transplund=pd.merge(codelund, restlund, on='pmid')

transplund['Institution']='Lund'
```

```python
codeki=pd.read_csv('./output/codesharing_ki.csv')
restki=pd.read_csv('./output/resttransparency_ki.csv')

transpki=pd.merge(codeki, restki, on='pmid')

transpki['Institution']='Karolinska'

transpki=transpki.drop_duplicates(subset=['pmid'])
```

```python
codestanford=pd.read_csv('./output/codesharing_stanford.csv')
reststanford=pd.read_csv('./output/resttransparency_stanford.csv')

transpstanford=pd.merge(codestanford, reststanford, on='pmid')

transpstanford['Institution']='Stanford'
```

```python
transplund = transplund.rename(columns={'pmid':'PMID'})
transpki = transpki.rename(columns={'pmid':'PMID'})
transpstanford = transpstanford.rename(columns={'pmid':'PMID'})
```

```python
transpki['is_open_code'] = transpki['is_open_code'].astype('bool')
transpki['is_open_data'] = transpki['is_open_data'].astype('bool')

transplund['is_open_code'] = transplund['is_open_code'].astype('bool')
transplund['is_open_data'] = transplund['is_open_data'].astype('bool')

transpstanford['is_open_code'] = transpstanford['is_open_code'].astype('bool')
transpstanford['is_open_data'] = transpstanford['is_open_data'].astype('bool')
```

```python
pmcoaki=pd.read_csv('./pmcoalist_ki.csv')
pmcoalund=pd.read_csv('./pmcoalist_lund.csv')
pmcoastanford=pd.read_csv('./pmcoalist_stanford.csv')
```

```python
transpki=pd.merge(transpki,pmcoaki,on='PMID')
transplund=pd.merge(transplund,pmcoalund,on='PMID')
transpstanford=pd.merge(transpstanford,pmcoaki,on='PMID')

# the ones that are research
transpkiresearch=transpki[transpki.is_research_x]
transplundresearch=transplund[transplund.is_research_x]
transpstanfordresearch=transpstanford[transpstanford.is_research_x]
```

```python
transpstanford['Publication_Year']=transpstanford['Publication Year']
columns=['PMID','is_coi_pred','is_register_pred','is_fund_pred','is_open_code','is_open_data','Publication_Year','Institution']
transps=pd.DataFrame(transpstanford, columns=columns)
transps['is_open_code'] = transps['is_open_code'].astype('bool')
transps['is_open_data'] = transps['is_open_data'].astype('bool')
```

```python
transpki['Publication_Year']=transpki['Publication Year']
columns=['PMID','is_coi_pred','is_register_pred','is_fund_pred','is_open_code','is_open_data','Publication_Year','Institution']
transpk=pd.DataFrame(transpki, columns=columns)
transpk['is_open_code'] = transpk['is_open_code'].astype('bool')
transpk['is_open_data'] = transpk['is_open_data'].astype('bool')
```

```python
transplund['Publication_Year']=transplund['Publication Year']
columns=['PMID','is_coi_pred','is_register_pred','is_fund_pred','is_open_code','is_open_data','Publication_Year','Institution']
transpl=pd.DataFrame(transplund, columns=columns)
transpl['is_open_code'] = transpl['is_open_code'].astype('bool')
transpl['is_open_data'] = transpl['is_open_data'].astype('bool')
```

```python
transps.to_csv('stanford.csv')
transpk.to_csv('ki.csv')
transpl.to_csv('lund.csv')
```

```python
all=pd.concat([transpk, transps])
all=pd.concat([all,transpl])

all.to_csv('all.csv')
```
